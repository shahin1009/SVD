```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10, 10)
import numpy as np
from PIL import Image
```

```python
# Define the Data
x = 3 # the slope of the line
a = np.arange(-2,2,0.25) # Create a vector of values from -2 to 2 with 0.25 steps
a = a.reshape(-1, 1) # Reshape the vector to a matrix
b = x*a + np.random.randn(*a.shape) # add noise to the data

plt.plot(a, x*a, color='k', linestyle='--') # Plot the line y = x
plt.plot(a, b, 'x', color='r', label='Data points') # Plot the data points

U, S, VT = np.linalg.svd(a,full_matrices=False) # Compute the SVD
xtilde = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ b # Compute the projection: the regression line

plt.plot(a,xtilde * a,'--',color='y',linewidth=4, label='Regression line') # Plot the regression line

plt.xlabel('a') # Label the x-axis
plt.ylabel('b') # Label the y-axis

plt.grid(linestyle='--') # Add a grid
plt.legend() 
plt.show()
```

# Second Method for computing psudo inverse of a matrix

```python
xtilde2 = np.linalg.pinv(a) @ b # Compute the projection: the regression line with numpy pseudo-inverse built-in function
plt.plot(a, b, 'x', color='r', label='Data points') # Plot the data points
plt.plot(a,xtilde2 * a,'--',color='y',linewidth=4, label='Regression line') # Plot the regression line
plt.xlabel('a') # Label the x-axis
plt.ylabel('b') # Label the y-axis
plt.grid(linestyle='--') # Add a grid
plt.legend()
plt.show()
```
